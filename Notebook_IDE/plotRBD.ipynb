```python
import matplotlib.pyplot as plt
import numpy as np
from math import exp

MTTF = 1400

t = np.linspace(0,MTTF*2)

lam = 1/MTTF
m = 5
s = 3
y1 = 1-(  1-np.exp(-s*lam*t)   )**m
y2 = (1-(1-np.exp(-lam*t))**m  )**s

# setting the axes at the centre
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
plt.plot(t,y1, 'b')
plt.plot(t,y2, 'r')

# show the plot
plt.show()

MT = 1100 #COSI ABBIAMOD DECISO
r_mt = np.exp(-lam*MT)
m1 = np.log(   (   1 - ( 1 - (1-r_mt)**m )**s   )  )  / np.log( 1-r_mt**s  )
print(m1)

t = np.linspace(0,MTTF*2)

m1 = 20
s = 3
y1 = 1-(  1-np.exp(-s*lam*t)   )**m1
y2 = (1-(1-np.exp(-lam*t))**m  )**s

# setting the axes at the centre
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
plt.plot(t,y1, 'b')
plt.plot(t,y2, 'r')

# show the plot
plt.show()
```
